```python
smalldfs = []
for df in df_all:
    df.drop(columns=todrop, errors='ignore', inplace=True)
    if 'vin model year' in df.columns:
        if 'model year' in df.columns:
            df.drop(columns='vin model year', errors='ignore', inplace=True)
        else:
            df['model year'] = df['vin model year'].apply(convert_vin)
            df.drop(columns='vin model year', errors='ignore', inplace=True)
    df.rename(columns={'stcountyfp':'county geoid', 'countyname':'county'}, errors='ignore', inplace=True)
    if 'county' in df.columns and 'county geoid' not in df.columns:
        df['county'] = df['county'].apply(lambda s: s.replace('County', '').strip()+' County')
        df = df.merge(zipdata[['stcountyfp', 'countyname']], how='left', left_on='county', right_on='countyname').drop(columns='countyname')
#     elif 'county geoid' in df.columns and 'county' not in df.columns:
# #         try:
#         df = df[df['county geoid'] != 'Unknown']
#         df = df.merge(zipdata[['stcountyfp', 'countyname']], how='left', left_on='county geoid', right_on='stcountyfp').drop(columns='stcountyfp')
# #         except:
# #             df = df[df['county geoid'] != 'Unknown']
# #             df['county geoid'] = df['county geoid'].astype('int64')
# #             df = df.merge(zipdata[['stcountyfp', 'countyname']], how='left', left_on='county geoid', right_on='stcountyfp').drop(columns='stcountyfp')
    df['registration year'] = pd.to_datetime(df['registration valid date']).dt.year
    df.rename(columns={'stcountyfp':'county geoid', 'countyname':'county'}, errors='ignore', inplace=True)
    df.drop(columns='registration valid date', inplace=True, errors='ignore')
    cols = list(df.columns)
    cols.sort()
    df = df[cols]
    display(df.head(1))
    smalldfs.append(df)
```

```python
grouped_dfs = []
for df in smalldfs:
    df['count'] = np.repeat('1', len(df))
    funcdict = {c:'first' for c in df.columns if c not in ['registration year', 'county geoid']}
    funcdict['count'] = 'count'
    grouped = df.groupby(['registration year', 'county geoid']).aggregate(funcdict).reset_index()
    grouped_dfs.append(grouped)
    display(grouped.head())
```

```python
for df in grouped_dfs:
    if 'state' in df.columns and 'state abbreviation' in df.columns:
        df.drop(columns=['state'], inplace=True)
    df.rename(columns={'state abbreviation':'state'}, inplace=True, errors='ignore')
    display(df.head(1))
```

```python
for df in grouped_dfs:
    display(df.head(1))
```

```python
grouped_final = []
for df in grouped_dfs:
    if 'county' not in df.columns:
        df = df.merge(zipdata[['stcountyfp', 'countyname']], how='left', left_on='county geoid', right_on='stcountyfp').drop(columns='stcountyfp')
        df = df.rename(columns={'countyname':'county'})
    cols = list(df.columns)
    cols.sort()
    df = df[cols]
    grouped_final.append(df)
        
```

```python
for df in grouped_final:
    display(df.head(1))
```

```python
vehicles_all = pd.concat(grouped_final, ignore_index=True)
```
